```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
J = 0.5 #Background dep rate
B = 2 #Base Instrinsic Growth rate
t = 10 #generation time
C = np.random.rand(10,10) #Trait — how burial stimulates growth

#https://www.wolframalpha.com/input/?i=ρ%27%28t%29%3D+%28B*ρ%28t%29%29+-++%28B*ρ%28t%29*ρ%28t%29%29%2B%28C*J*ρ%28t%29%29%3B+ρ%280%29%3D0.01
ρ = ((B + (C*J))*np.exp(t*(B + (C*J))))/(B*np.exp(t*(B + (C*J))) + (99*B) + (100*C*J))

c = plt.pcolor(ρ)
plt.colorbar()
plt.show()
```
